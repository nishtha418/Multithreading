```python
import os, sys, time, threading, multiprocessing, random as r

# Function runs for random seconds
def task(cmd):

    waitingTime = r.randint(5,10)

    print("  Task started")
    print("   Task running for %d sec...." % (waitingTime))
    time.sleep(waitingTime)

    print("  Task ends")
    return



# Main Program
startTime = time.time()

cmd = 'Sample task'

print("Program Started")
print(" Thread starts")

t = threading.Thread(target=task , args=(cmd,))
t.start()

print(" Thread ends")

print("Program Finished")
print("Total Time %f sec" % (round( time.time() - startTime,4)))
```

**3. waits for the thread to complete**

```python
import os, sys, time, threading, multiprocessing, random as r

# Function runs for random seconds
def task(cmd):

    waitingTime = r.randint(5,10)

    print("  Task started")
    print("   Task running for %d sec...." % (waitingTime))
    time.sleep(waitingTime)

    print("  Task ends")
    return



# Main Program
startTime = time.time()

activeThreads = threading.activeCount()
print("Active Threads = ", activeThreads)

cmd = 'Sample task'


print("Program Started")
print(" Thread starts")

t = threading.Thread(target=task , args=(cmd,))
t.start()
time.sleep(1)

# Waiting to finish the Threads
while True:
    if threading.activeCount() == activeThreads:
        break
    else:
        print ("    Thread still running (left %d)..."%(threading.activeCount() - activeThreads))
        time.sleep(1)

print(" Thread ends")

print("Program Finished")
print("Total Time %f sec" % (round( time.time() - startTime,4)))
```

**4. Run multiple threads and waits to complete all **

```python
import os, sys, time, threading, multiprocessing, random as r

# Function runs for random seconds
def task(cmd):

    waitingTime = r.randint(5,10)

    print("  Task started")
    print("   Task running for %d sec...." % (waitingTime))
    time.sleep(waitingTime)

    print("  Task ends")
    return


# Main Program - Run 16 threads in one go

startTime = time.time()

numberOfThreads = 16

# Main Program - Run Multiple Threads
print("Program Started....")

for i in range(numberOfThreads):
    cmd = str(i+1)
    print(" Thread %s starts"%(cmd))
    t = threading.Thread(target=task , args=(cmd,))
    t.start()

time.sleep(1)

# Waiting to finish all Threads
while True:
    if threading.activeCount() == activeThreads:
        break
    else:
        print ("    Thread still running (left %d)..."%(threading.activeCount() - activeThreads))
        time.sleep(1)

print(" All Thread ends")

print("Program Finished")
print("Total Time %f sec" % (round( time.time() - startTime,4)))
```

**Run no of threads = no of cores**

```python
import os, sys, time, threading, multiprocessing, random as r

# Function runs for random seconds
def task(cmd):

    waitingTime = r.randint(5,10)

    print("  Task started")
    print("   Task running for %d sec...." % (waitingTime))
    time.sleep(waitingTime)

    print("  Task ends")
    return


# Main Program

startTime = time.time()

numberOfThreads = 16

activeThreads = threading.activeCount()
print("Active Threads = ", activeThreads)

numberOfCores=multiprocessing.cpu_count()
print("Number Of Cores = ", numberOfCores)

# Main Program - Run Multiple Threads
print("Program Started....")

for i in range(numberOfThreads):
    cmd = str(i+1)
    print(" Thread %s starts"%(cmd))
    t = threading.Thread(target=task , args=(cmd,))
    t.start()
    while True:
      if threading.activeCount() - activeThreads + 1 <= numberOfCores:
        break
      time.sleep(1)

time.sleep(1)

# Waiting to finish all Threads
while True:
    if threading.activeCount() == activeThreads:
        break
    else:
        print ("    Thread still running (left %d)..."%(threading.activeCount() - activeThreads))
        time.sleep(1)

print(" All Thread ends")

print("Program Finished")
print("Total Time %f sec" % (round( time.time() - startTime,4)))
```

**Project-1: Create multiple files using threading***

** => create multiple files one by one **

```python
import time
import os

def create_file(i):
    with open(f"file_{i}.txt", "w") as f:
        f.write("Hello from single thread\n")

def run(n):
    start = time.time()

    for i in range(n):
        create_file(i)

    end = time.time()
    print(f"Single Thread Time: {round(end - start, 4)} sec")
    return round(end - start, 4)

if __name__ == "__main__":
    run(100)
```

**=> create multiple files using multithreading **

```python
import time
import threading

def create_file(i):
    with open(f"file_{i}.txt", "w") as f:
        f.write("Hello from thread\n")

def run(n):
    threads = []
    start = time.time()

    for i in range(n):
        t = threading.Thread(target=create_file, args=(i,))
        t.start()
        threads.append(t)

    # Wait for all threads
    for t in threads:
        t.join()

    end = time.time()
    print(f"Multi Thread Time: {round(end - start, 4)} sec")
    return round(end - start, 4)

if __name__ == "__main__":
    run(100)
```

**compare single thread vs multithread**

```python
import time
import threading
import matplotlib.pyplot as plt

# -------- SINGLE THREAD --------
def single_run(n):
    def create_file(i):
        with open(f"file_{i}.txt", "w") as f:
            f.write("Single thread\n")

    start = time.time()
    for i in range(n):
        create_file(i)
    return round(time.time() - start, 4)

# -------- MULTI THREAD --------
def multi_run(n):
    def create_file(i):
        with open(f"file_{i}.txt", "w") as f:
            f.write("Multi thread\n")

    threads = []
    start = time.time()

    for i in range(n):
        t = threading.Thread(target=create_file, args=(i,))
        t.start()
        threads.append(t)

    for t in threads:
        t.join()

    return round(time.time() - start, 4)

# -------- COMPARISON --------
file_counts = [10, 50, 100]

single_times = []
multi_times = []

for n in file_counts:
    print(f"\nRunning for {n} files...")

    t1 = single_run(n)
    t2 = multi_run(n)

    single_times.append(t1)
    multi_times.append(t2)

# -------- GRAPH --------
plt.plot(file_counts, single_times, marker='o', label="Single Thread")
plt.plot(file_counts, multi_times, marker='o', label="Multi Thread")

plt.xlabel("Number of Files")
plt.ylabel("Time (seconds)")
plt.title("Single vs Multi Thread Performance")
plt.legend()

plt.show()
```

```python
import time
import os
import threading
from PIL import Image

def process_image(filename):
    img = Image.open(f"input_images/{filename}")
    gray = img.convert("L")
    gray.save(f"output_multi/{filename}")

def run():
    threads = []
    start = time.time()

    for file in os.listdir("input_images"):
        t = threading.Thread(target=process_image, args=(file,))
        t.start()
        threads.append(t)

    for t in threads:
        t.join()

    end = time.time()
    print("Multi Thread Time:", round(end - start, 4))
    return round(end - start, 4)
```
